```python
import matplotlib.pyplot as plt
```

```python

year = ['2019', '2020', '2021', '2022', '2023']
allpeo= [51849861, 51829023, 51638809, 51439038, 51325329]
foreign= [2524656, 2036075, 1956781, 2245912, 2507584]

li=[]
for i in range(int(len(allpeo))):
    li.append(foreign[i]/allpeo[i]*100)

li
```

```python
import seaborn as sns
import koreanize_matplotlib

colors = sns.color_palette('pastel',len(year))

plt.bar(year, li, color=colors)
plt.ylabel('%\n(Per)', rotation=0, labelpad=30)
plt.ylim(3, 6)
plt.title('연도별 인구대비 외국인 비율')
```
