```python
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
```

```python
# 드라이버 옵션 설정

chrome_options = webdriver.ChromeOptions()
chrome_options.add_experimental_option("detach", True) # 브라우저 열어놓을지 선택

# 크롬 매니저 설치
driver = webdriver.Chrome(service = Service(ChromeDriverManager().install(),
                          options=chrome_options))

# 3.웹드라이버 주소 넣기

url = "https://gall.dcinside.com/board/lists/?id=wedding"
driver.get(url)
time.sleep(2) # 로딩 기다리는 시간

# 개념글 게시물로 이동
driver.find_element(By.XPATH, '//*[@id="container"]/section[1]/article[2]/div[1]/div[1]/button[2]').click()

###########################        개념글 페이지        ###############################
```

```python
# 게시글 제목 + link
def title_link(title, link):
    data = driver.find_elements(By.CLASS_NAME, 'ub-word')

    # link 따기
    for i in data[2:]: # 공지글 제거 
        a=i.find_element(By.TAG_NAME, 'a')
        link.append(a.get_attribute('href'))
    
    # 제목따기
    for i in data[2:]:
        title.append(i.text)
    
```

```python
title = []
link = []

for num in range(1,20+1):
    driver.find_element(By.XPATH, '//*[@id="container"]/section[1]/article[2]/div[4]/div[2]/button').click()
    driver.find_element(By.XPATH, '//*[@id="container"]/section[1]/article[2]/div[4]/div[3]/div/div[2]/div[2]/input').click()
    driver.find_element(By.XPATH, '//*[@id="container"]/section[1]/article[2]/div[4]/div[3]/div/div[2]/div[2]/input').clear()
    driver.find_element(By.XPATH, '//*[@id="container"]/section[1]/article[2]/div[4]/div[3]/div/div[2]/div[2]/input').send_keys(int(num))
    driver.find_element(By.XPATH, '//*[@id="container"]/section[1]/article[2]/div[4]/div[3]/div/div[2]/div[2]/button').click()
    title_link(title, link) # 제목과 링크 가져오기
    print(f'{num}page complite')

```

```python
print(len(title), len(link))
```

```python
contents = []
comments = []
num = 1
for i in link[0:100]:
    driver.get(i)

    # 게시글 내용 복사
    contents.append(driver.find_element(By.CLASS_NAME, 'write_div').text)

    # 게시글 댓글 복사
    for i in driver.find_elements(By.CLASS_NAME, 'ub-word'):
        comments.append(i.text)

    print(f'{num}/{len(link)} 페이지 완료')
    num+=1
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np
from konlpy.tag import Okt
from konlpy.tag import Mecab 
import re # 특수 문자 등 필요 없는 문자 제거를 위한 라이브러리
```

```python
okt = Okt()
con = "".join(con) # 문자열로 변경
con = re.sub(r'[^가-힣\s]', '', con) # 한국어를 제외한 문자 제거
```

```python
mecab= Mecab(dicpath='C:\mecab\mecab-ko-dic') # 명사 추출 객체
con_nouns = mecab.nouns(con)
con_nouns
```

```python
# mecab 처리 텍스트
stopwords = ['저', '때', '당시', '지금', '처녀', '자지', '난자', '정자', '섹스', '프리 섹스']

# 불용어 제거
con_nouns = [word for word in con_nouns if word not in stopwords]
con_nouns
```

```python
result = []
for i in con_nouns:
    if len(i) > 1:
        result.append(i)
result = " ".join(result)
```

```python
# 워드 클라우드 객체 생성
wordcloud = WordCloud(
    font_path='C:/Windows/Fonts/malgun.ttf',  # 한글 폰트 경로
    background_color='white',  # 배경색
    max_words=200,  # 최대 단어 수
    width=800, height=400  # 크기
).generate(result)

# 워드 클라우드 시각화
plt.figure(figsize=(8, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # 축 제거
plt.show()
```
