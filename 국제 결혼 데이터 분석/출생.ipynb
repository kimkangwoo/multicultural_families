```python
import folium
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import koreanize_matplotlib
```

```python
f = pd.read_csv('./지역별_출생아_수.csv')
file = f.iloc[:, :-2]
file.info()
```

```python
# 세종의 2011년 데이터가 '-'로 표기
# 세종의 평균값으로 변환

file.iloc[8, 1]=int(file.iloc[8, 2:].min())
file.info()
```

```python
# 2012년 데이터가 object 타입으로 int로 변환
file['2011'] = file['2011'].astype(int)
file.info()
```
